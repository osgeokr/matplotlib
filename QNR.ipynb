```python
# 깃대종 인지도 설문
# 문1-1. (문1에서 ‘예’를 선택한 경우) ‘깃대종’하면 무엇이 가장 먼저 떠오르시나요?

# 꼬꼬마 형태소 분석기
from konlpy.tag import Kkma
kkma = Kkma()

# 문1-1
df[['F1-1']].to_csv('D:/GEODATA/QNR_F1-1.txt', index=False, header=False)
f = open("D:/GEODATA/QNR_F1-1.txt", 'r', encoding='UTF-8')
lines = f.readlines()
f.close()

# 단어 2차원 리스트
word = []
for i in range(len(lines)):
    word.append(kkma.nouns(lines[i]))
word = [[y for y in x if not len(y)==1] for x in word]   # 2자 이상 
word = [[y for y in x if not y.isdigit()] for x in word] # 숫자판별

# 단어 1차원 리스트 함수
def flatten(l):
    flatList = []
    for elem in l:
        if type(elem) == list:
            for e in elem:
                flatList.append(e)
        else:
            flatList.append(elem)
    return flatList

# 키워드 단어구름
from wordcloud import WordCloud
from collections import Counter

font_path = 'c:/Windows/Fonts/KoPubDotumBold.ttf'
wordcloud = WordCloud(font_path=font_path, width=800, height=800, background_color="white")

count = Counter(flatten(word))

wordcloud = wordcloud.generate_from_frequencies(count)
array = wordcloud.to_array()

fig = plt.figure(figsize=(10,10))
plt.imshow(array, interpolation="bilinear")
plt.title('문1-1. (문1에서 ‘예’를 선택한 경우) ‘깃대종’하면 무엇이 가장 먼저 떠오르시나요?')
plt.axis("off")
plt.savefig('D:/GEODATA/IMG15.png', bbox_inches='tight')
plt.show()
```

```python
# 깃대종 인지도 설문
# 문10. 끝으로 깃대종 대국민 인지도 향상에 있어 중요한 점이 있다면 자유롭게 말씀해주시기 바랍니다.

# 꼬꼬마 형태소 분석기
from konlpy.tag import Kkma
kkma = Kkma()

# 문1-1
df[['F10']].to_csv('D:/GEODATA/QNR_F10.txt', index=False, header=False)
f = open("D:/GEODATA/QNR_F10.txt", 'r', encoding='UTF-8')
lines = f.readlines()
f.close()

# 단어 2차원 리스트
word = []
for i in range(len(lines)):
    word.append(kkma.nouns(lines[i]))
word = [[y for y in x if not len(y)==1] for x in word]   # 2자 이상 
word = [[y for y in x if not y.isdigit()] for x in word] # 숫자판별

# 단어 1차원 리스트 함수
def flatten(l):
    flatList = []
    for elem in l:
        if type(elem) == list:
            for e in elem:
                flatList.append(e)
        else:
            flatList.append(elem)
    return flatList

# 키워드 단어구름
from wordcloud import WordCloud
from collections import Counter

font_path = 'c:/Windows/Fonts/KoPubDotumBold.ttf'
wordcloud = WordCloud(font_path=font_path, width=800, height=800, background_color="white")

count = Counter(flatten(word))

stopwords = ["깃대종", "깃대", "깃대종이", "국립공원", "종이"]
[count.pop(stopword) for stopword in stopwords]

wordcloud = wordcloud.generate_from_frequencies(count)
array = wordcloud.to_array()

fig = plt.figure(figsize=(10,10))
plt.imshow(array, interpolation="bilinear")
plt.title('문10. 끝으로 깃대종 대국민 인지도 향상에 있어 중요한 점이 있다면 자유롭게 말씀해주시기 바랍니다.')
plt.axis("off")
plt.savefig('D:/GEODATA/IMG24.png', bbox_inches='tight')
plt.show()
```
