```python
# 라이브러리 추가
import pandas as pd
import numpy as np
import seaborn as sns; sns.set(font_scale=1.5)
import matplotlib.pyplot as plt

# 한글 폰트 설정
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/KoPubDotumBold.ttf").get_name()
rc('font', family=font_name, size=20)
```

```python
# '국립공원' 열지도
df1 = pd.read_excel('D:/GEODATA/datalab.xlsx', sheet_name='깃대종' ,index_col='날짜', parse_dates=True)
df1['Year'] = df1.index.year; df1['Month'] = df1.index.month
df1 = df1.pivot("Month", "Year", "국립공원")

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(16, 10))
sns.heatmap(df1, annot=True, fmt='.1f', linewidths=.5, ax=ax)
plt.savefig('D:/GEODATA/NAVERAPI3_IMG01.png', bbox_inches='tight')
```

```python
# '깃대종' 열지도
df2 = pd.read_excel('D:/GEODATA/datalab.xlsx', sheet_name='깃대종' ,index_col='날짜', parse_dates=True)
df2['Year'] = df2.index.year; df2['Month'] = df2.index.month
df2 = df2.pivot("Month", "Year", "깃대종")

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(16, 10))
sns.heatmap(df2, annot=True, fmt='.1f', linewidths=.5, ax=ax)
plt.savefig('D:/GEODATA/NAVERAPI3_IMG02.png', bbox_inches='tight')
```

```python
# '국립공원'과 '깃대종'의 열지도 비교
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 10))
sns.heatmap(df1, annot=True, fmt='.1f', linewidths=.5, ax=ax1)
sns.heatmap(df2, annot=True, fmt='.1f', linewidths=.5, ax=ax2)
ax1.set_title('국립공원 검색빈도')
ax2.set_title('깃대종 검색빈도')
plt.savefig('D:/GEODATA/NAVERAPI3_IMG03.png', bbox_inches='tight')
plt.show()
```

```python
# '국립공원'과 '깃대종'의 열지도 비교 (색상지도 변경)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 10))
sns.heatmap(df1, cmap="YlGnBu", annot=True, fmt='.1f', linewidths=.5, ax=ax1)
sns.heatmap(df2, cmap="YlGnBu", annot=True, fmt='.1f', linewidths=.5, ax=ax2)
ax1.set_title('국립공원 검색빈도')
ax2.set_title('깃대종 검색빈도')
plt.savefig('D:/GEODATA/NAVERAPI3_IMG04.png', bbox_inches='tight')
plt.show()
```
