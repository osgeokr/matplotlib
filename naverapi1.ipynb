```python
# 네이버 블로그 검색 (최대 100개)
keyword = "깃대종"
url = "https://openapi.naver.com/v1/search/blog?query=" + keyword + "&display=100"
result = requests.get(urlparse(url).geturl(),
                      headers={"X-Naver-Client-Id":"89L30Y3xCshEctCoab1v", "X-Naver-Client-Secret":"aIoL3WXAsj"})
json_obj = result.json()
for item in json_obj['items']:
    print(item['title'].replace("<b>","").replace("</b>",""),
          item['description'].replace("<b>","").replace("</b>",""), item['postdate'], item['link'])
```

```python
# 네이버 블로그 검색 (keyword: 검색어, display: 검색결과 출력 건수, start: 검색 시작 위치; 최대 1000까지 가능)
def get_api_result(keyword, display, start):
    url = "https://openapi.naver.com/v1/search/blog?query=" + keyword + "&display=" + str(display) + "&start=" + str(start)
    result = requests.get(urlparse(url).geturl(),
                          headers={"X-Naver-Client-Id":"89L30Y3xCshEctCoab1v", "X-Naver-Client-Secret":"aIoL3WXAsj"})
    return result.json()

list = []
def call_and_print(keyword, page):
    json_obj = get_api_result(keyword, 100, page)
    for item in json_obj['items']:
        val = [item['title'].replace("<b>","").replace("</b>",""),
               item['description'].replace("<b>","").replace("</b>",""), item['postdate'], item['link']]
        list.append(val)
```

```python
# 깃대종 블로그 검색
keyword = "깃대종"
page_list = [1, 101, 201, 301, 401, 501, 601, 701, 801, 901]
for i in page_list:
    call_and_print(keyword, i)
```

```python
# 깃대종 블로그 검색결과 데이터프레임 생성
df = pd.DataFrame(list, columns = ['title', 'description', 'postdate', 'link'])
df.head(10)
```

```python
# 깃대종 블로그 검색결과 엑셀파일로 저장
df.to_excel('D:/GEODATA/naverblog.xlsx')
```

```python
# 엑셀파일 읽기
data = pd.read_excel('D:/GEODATA/naverblog.xlsx', sheet_name='Sheet1', index_col='postdate', parse_dates=True)
```

```python
# 기간별 기사 건수
df1 = data.groupby(data.index.year).size()
df1.plot(kind='bar')
plt.xlabel('기간')
plt.ylabel('블로그글 건수')
plt.title('기간별 깃대종 블로그글 건수')
plt.savefig('D:/GEODATA/NAVERAPI1_IMG01.png', bbox_inches='tight')
plt.show()
```

```python
# 기간별 기사 건수
fig, ax = plt.subplots()
plt.xticks(df1.index, rotation=90)
plot = ax.bar(df1.index, df1.values)
              
# 라벨 추가
for rect in plot:
              height = rect.get_height()
              ax.text(rect.get_x() + rect.get_width()/2., 1.002*height, '%d' %int(height), ha='center', va='bottom')
plt.xlabel('기간')
plt.ylabel('블로그글 건수')
plt.title('기간별 깃대종 블로그글 건수')
plt.savefig('D:/GEODATA/NAVERAPI1_IMG02.png', bbox_inches='tight')
plt.show()
```

```python
# '국립공원'이 언급된 블로그글 건수
df2 = data[data['title'].str.contains('국립공원') | data['description'].str.contains('국립공원')]
df2 = df2.groupby(df2.index.year).size()

# 기간별 기사 건수
fig, ax = plt.subplots()
plt.xticks(df2.index, rotation=90)
plot = ax.bar(df2.index, df2.values)
              
# 라벨 추가
for rect in plot:
              height = rect.get_height()
              ax.text(rect.get_x() + rect.get_width()/2., 1.002*height, '%d' %int(height), ha='center', va='bottom')
plt.xlabel('기간')
plt.ylabel('블로그글 건수')
plt.title('국립공원이 언급된 기간별 깃대종 블로그글 건수')
plt.savefig('D:/GEODATA/NAVERAPI1_IMG03.png', bbox_inches='tight')
plt.show()
```

```python
# 전체 깃대종과 국립공원이 언급된 깃대종 블로그글 건수
pd.concat([df1, df2], axis=1).plot(kind='bar')
plt.legend(["전체 깃대종 블로그글 건수", "국립공원이 언급된 깃대종 블로그글 건수"])
plt.xlabel('기간')
plt.ylabel('블로그글 건수')
plt.title('기간별 깃대종 블로그글 건수')
plt.savefig('D:/GEODATA/NAVERAPI1_IMG04.png', bbox_inches='tight')
plt.show()
```

```python
# 전체 깃대종 블로그글 중 국립공원이 언급된 블로그글의 비중
explode = (0, 0.1)
plt.pie(pd.concat([df1, df2], axis=1).sum(), explode=explode, autopct='%1.1f%%')
plt.title('기간별 깃대종 블로그글 건수')
plt.legend(["전체 깃대종 블로그글 건수", "국립공원이 언급된 깃대종 블로그글 건수"], loc='lower right', bbox_to_anchor=(1.3, 0))
plt.savefig('D:/GEODATA/NAVERAPI1_IMG05.png', bbox_inches='tight')
plt.show()
```
