```python
# 라이브러리 추가
import datetime as dt
import pandas as pd

from twitterscraper import query_tweets
```

```python
# 트위터 검색
list = []
list_of_tweets = query_tweets('깃대종', begindate=dt.date(2006,1,1), enddate=dt.date(2018,12,31), lang='ko')
for tweet in list_of_tweets:
    if '깃대종' in tweet.text:
        val = [tweet.fullname, tweet.username, tweet.text, tweet.timestamp, tweet.tweet_url]
        list.append(val)

df = pd.DataFrame(list, columns = ['fullname', 'username', 'text', 'timestamp', 'url'])
df.head(10)
```

```python
# 트위터 검색결과 엑셀파일로 저장
df.to_excel('D:/GEODATA/twitter.xlsx')
```
