```python
# 네이버 카페 검색 (최대 100개)
keyword = "깃대종"
url = "https://openapi.naver.com/v1/search/cafearticle?query=" + keyword + "&display=100"
result = requests.get(urlparse(url).geturl(),
                      headers={"X-Naver-Client-Id":"89L30Y3xCshEctCoab1v", "X-Naver-Client-Secret":"aIoL3WXAsj"})
json_obj = result.json()
for item in json_obj['items']:
    print(item['title'].replace("<b>","").replace("</b>",""),
          item['description'].replace("<b>","").replace("</b>",""), item['cafename'], item['link'])
```

```python
# 네이버 카페 검색 (keyword: 검색어, display: 검색결과 출력 건수, start: 검색 시작 위치; 최대 1000까지 가능)
def get_api_result(keyword, display, start):
    url = "https://openapi.naver.com/v1/search/cafearticle?query=" + keyword + "&display=" + str(display) + "&start=" + str(start)
    result = requests.get(urlparse(url).geturl(),
                          headers={"X-Naver-Client-Id":"89L30Y3xCshEctCoab1v", "X-Naver-Client-Secret":"aIoL3WXAsj"})
    return result.json()

list = []
def call_and_print(keyword, page):
    json_obj = get_api_result(keyword, 100, page)
    for item in json_obj['items']:
        val = [item['title'].replace("<b>","").replace("</b>",""),
               item['description'].replace("<b>","").replace("</b>",""), item['cafename'], item['link']]
        list.append(val)
```

```python
# 네이버 카페 검색
keyword = "깃대종"
page_list = [1, 101, 201, 301, 401, 501, 601, 701, 801, 901]
for i in page_list:
    call_and_print(keyword, i)
```

```python
# 네이버 카페 검색결과 데이터프레임 생성
df = pd.DataFrame(list, columns = ['title', 'description', 'cafename', 'link'])
df.head(10)
```
