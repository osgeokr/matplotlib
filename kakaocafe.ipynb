```python
# 라이브러리 추가
import pandas as pd
import requests

from urllib.parse import urlparse
```

```python
# 카카오 카페글 검색
keyword = "깃대종"
url = "https://dapi.kakao.com/v2/search/cafe?&query=" + keyword
result = requests.get(urlparse(url).geturl(),
                      headers={"Authorization":"KakaoAK 4d25f7ae6c4cab6bc96c746280e403af"})
json_obj = result.json()
print(json_obj)
```

```python
# 카카오 카페글 정렬
list = []
for document in json_obj['documents']:
    val = [document['title'].replace("<b>","").replace("</b>",""),
          document['contents'].replace("<b>","").replace("</b>",""),
          document['cafename'], document['datetime'], document['url']]
    list.append(val)

df = pd.DataFrame(list, columns = ['title', 'contents', 'cafename', 'datetime', 'url'])
df.head(10)
```

```python
# 카카오 카페글 검색 (keyword: 검색어, size: 검색결과 출력 건수, page: 검색 시작 위치; 최대 16까지 가능)
def get_api_result(keyword, size, page):
    url = "https://dapi.kakao.com/v2/search/cafe?&query=" + keyword + "&size=" + str(size) + "&page=" + str(page)
    result = requests.get(urlparse(url).geturl(),
                          headers={"Authorization":"KakaoAK 4d25f7ae6c4cab6bc96c746280e403af"})
    return result.json()

list = []
def call_and_print(keyword, page):
    json_obj = get_api_result(keyword, 50, page)
    for document in json_obj['documents']:
        val = [document['title'].replace("<b>","").replace("</b>",""),
              document['contents'].replace("<b>","").replace("</b>",""),
              document['cafename'], document['datetime'], document['url']]
        list.append(val)
    return json_obj
```

```python
# 카카오 카페글 검색
keyword = "깃대종"
i = 1
while i <= 50:
    json_obj = call_and_print(keyword, i)
    i += 1
    if json_obj['meta']['is_end'] is True:
        break
```

```python
# 카카오 카페글 검색결과 데이터프레임 생성
df = pd.DataFrame(list, columns = ['title', 'contents', 'cafename', 'datetime', 'url'])
df.head(10)
```

```python
# 카카오 카페글 검색결과 엑셀파일로 저장
df.to_excel('D:/GEODATA/kakaocafe.xlsx')
```
